```python
# ordinal Encoding
from sklearn.preprocessing import OrdinalEncoder

dtp=pd.DataFrame({"size":["s","m","l","xl","xxl"]})
```

```python
dtp
```

```python
dp=[["s","m","l","xl","xxl"]]
```

```python
na=OrdinalEncoder(categories=dp)
```

```python
na.fit(dtp[["size"]])
```

```python
na.transform(dtp[["size"]])
```

```python
dtp["size_en"]=na.transform(dtp[["size"]])
```

```python
dtp
```

```python
# through map function
ord1={"s":5,"m":4,"l":3,"xl":2,"xxl":1}
```

```python
dtp["size_enn"]=dtp["size"].map(ord1)
```

```python
dtp
```
