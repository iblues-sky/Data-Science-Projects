```python
import pandas as pd
import seaborn as  sns
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
```

```python
df = pd.read_csv("placement-dataset.csv")
df.head(3)
```

```python
dff = df[1:]
```

```python
dff.head(3)
```

```python
sns.kdeplot(data = dff["cgpa"])
plt.show()
```

```python
sns.kdeplot(data = dff["iq"])
plt.show()
```

```python
sns.scatterplot(x="cgpa",y="iq",data=dff,hue="placement")
plt.show()
```

```python
x = dff.iloc[:,:-1]
y = dff["placement"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

### We 3 Techniques in Naive bayes Theorm  1. Gaussian we used this when our data foloow Normal Distribuution

```python
from sklearn.naive_bayes import GaussianNB,BernoulliNB,MultinomialNB
```

```python
Gu = GaussianNB()
Gu.fit(x_train,y_train)
```

```python
Gu.score(x_test,y_test)*100,Gu.score(x_train,y_train)*100
```

### In This Chart We have some Miss classification But over all its good By Gaussian method

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=Gu)
plt.show()
```

### 2. Bernaulli Distribution we used this when our data are in the form of True or False OR 0 or 1

```python
Br = BernoulliNB()
Br.fit(x_train,y_train)
```

```python
Br.score(x_test,y_test)*100,Br.score(x_train,y_train)*100
```

### In This Chart We have  Miss classification By Bernoulli method

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=Br)
plt.show()
```

### 3. Multinomail Distribution We used this for Text classification

```python
Mu = MultinomialNB()
Mu.fit(x_train,y_train)
```

```python
Mu.score(x_test,y_test)*100,Mu.score(x_train,y_train)*100
```

### In This Chart We have Miss classification  By Multinomail method

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=Mu)
plt.show()
```
