```python
import pandas as pd
```

```python
df=pd.read_csv("aug_train.csv")
df.head(3)
```

```python
df.isnull().sum()
```

```python
# fill any value in missing place by using this function
# but its not a right way because in both categorical and numerical data fills any data which is wrong
df.fillna(10)
```

```python
df.isnull().sum()
```

```python
# backward filling technique for filling missing data
df.fillna(method="bfill")
```

```python
# forward filling technique for filling missing data
df.fillna(method="ffill")
```

```python
# fill mode in missing content
df["major_discipline"].fillna((df["major_discipline"].mode()[0]))
```

```python
# Select all object types columns
df.select_dtypes(include="object")
```

```python
df.select_dtypes(include="object").isnull().sum()
```

```python
df.select_dtypes(include="object").columns
```

```python
for i in df.select_dtypes(include="object").columns:
    print(i)
```

```python
# filling mode by using loops
for i in df.select_dtypes(include="object").columns:
    df[i].fillna(df[i].mode()[0])
```
