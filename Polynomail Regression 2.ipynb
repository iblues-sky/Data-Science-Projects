```python
# Simple Logisticregression on polynomial datset
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("nonlinear_data_circles.csv")
```

```python
df.head(3)
```

```python
sns.scatterplot(x="x1",y="x2",data=df,hue="label")
plt.show()
```

```python
x = df.iloc[:,:-1]
y= df["label"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[0.160594,0.482602]])
```

```python

from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=lr)
plt.show()
```

```python
dp.head(3)
```

```python
sns.scatterplot(x="x1",y="x2",data=dp,hue="label")
plt.show()
```

```python
p = df.iloc[:,:-1]
s= df["label"]
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pf = PolynomialFeatures(degree=2)
```

```python
pf.fit(p)
p = pd.DataFrame(pf.transform(p))
```

```python
from sklearn.model_selection import train_test_split
```

```python
xp_train,xs_test,yp_train,ys_test = train_test_split(p,s,test_size = 0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lp = LogisticRegression()
```

```python
lp.fit(xp_train,yp_train)
```

```python
lp.score(xs_test,ys_test)*100
```
