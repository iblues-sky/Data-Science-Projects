```python
import pandas as pd
```

```python
df = pd.read_csv("Social_Network_Ads.csv")
df.head(3)
```

```python
df.drop(columns=["User ID","Gender"],inplace=True)
```

```python
df.head(10)
```

```python
x = df.iloc[:,:-1]
y = df["Purchased"]
```

```python
df["Purchased"].value_counts()
```

### Random Under sampler techniue

```python
from imblearn.under_sampling import RandomUnderSampler

```

```python
Ru = RandomUnderSampler()
```

```python
ru_x,ru_y = Ru.fit_resample(x,y)
```

```python
ru_y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(ru_x, ru_y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[32,150000]])
# good answer
```

```python
lr.predict([[35,65000]]) 
# Good answer
```

### Random Over sampler Technique

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
Ro = RandomOverSampler()
ro_x,ro_y = Ro.fit_resample(x,y)
```

```python
ro_y.value_counts()
```

```python
x_train1,x_test1,y_train1,y_test1 = train_test_split(ro_x, ro_y, test_size = 0.2, random_state = 42)
```

```python
lr1 = LogisticRegression()
```

```python
lr1.fit(x_train1,y_train1)
```

```python
lr1.score(x_test1,y_test1)*100
```

```python
lr1.predict([[32,150000]])
```

```python
lr1.predict([[35,65000]]) 
```
